```python
import bs4
import requests
import numpy as np
import pandas as pd
from bs4 import BeautifulSoup as soup
```

```python
for year in range(2019,2021):
    
    url = "https://paperview.spp-online.org/proceedings/issue/view/SPP-" + str(year)
    response = requests.get(url,headers={'User-Agent': 'Mozilla/5.0 (Macintosh; Intel Mac OS X 10_13_4) AppleWebKit/537.36 (KHTML, like Gecko) Chrome/35.0.1916.47 Safari/537.36'})
    page_soup = soup(response.content, "html.parser")
    
    titles = []
    authors = []
    codes = []
    links = []
    affiliations = []
    abstracts = []
    count = 1
    
    title = page_soup.findAll("div", {"class":"title"}) #grab title info
    author = page_soup.findAll("div", {"class":"authors"}) #find author info
    code = page_soup.findAll("div", {"class":"pages"}) #article info
    
    for i in range(len(title)-1): #title, author, publication code collection
    
        ttxt = title[i].a.text
        ttxt = ttxt.replace('\t','').replace('\n','')
        titles.append(ttxt)
    
        atxt = author[i].text
        atxt = atxt.replace('\t','').replace('\n','')
        authors.append(atxt)  
    
        ctxt = code[i].text
        ctxt = ctxt.replace('\t','').replace('\n','')
        codes.append(ctxt)
    
        ltxt = title[i].a['href']
        links.append(ltxt)
    
    for link in links: #abstract and affiliation collection 
    
        url = link
        response = requests.get(url,headers={'User-Agent': 'Mozilla/5.0 (Macintosh; Intel Mac OS X 10_13_4) AppleWebKit/537.36 (KHTML, like Gecko) Chrome/35.0.1916.47 Safari/537.36'})
        page_soup = soup(response.content, "html.parser")
        abstract = page_soup.findAll("div",{"class":"item abstract"})
        affiliation = page_soup.findAll("span",{"class":"affiliation"})
    
        try:
    
            if abstract[0].p != None:
                abstracts.append(abstract[0].p.text.replace("\xa0",""))
        
            else:
                abstracts.append(np.NaN)
            
            if len(affiliation) == 0:
                affiliations.append(np.NaN)
            
            elif len(affiliation) == 1:
                affiliations.append(affiliation[0].text.replace('\n','').replace('\t',''))
            
            elif len(affiliation) > 1:
                aff = ''
                for i in affiliation:
                    aff = aff + i.text.replace('\n','').replace('\t','') + ", "
                affiliations.append(aff[0:len(aff)-2])
            
        except IndexError: #if webpage is mising (error 404)
            
            abstracts.append(np.NaN)
            affiliations.append(np.NaN)
        
        print(str(year)+": "+str(count)+" of "+str(len(links))) 
        count = count+1
        
    dic = {'author': authors, 'affiliation': affiliations, 'title': titles, 'abstract' : abstracts, 'code': codes}  
    df = pd.DataFrame(dic) 
    path = str(year)+"_spp_congress.csv"
    df.to_csv(path)
    print(str(year)+"_df is done!")
```
